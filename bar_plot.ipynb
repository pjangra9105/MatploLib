```python
# Imports
from matplotlib import pyplot as plt

# Bar Plot and Horizontal Bar Plot
    # Create a Dictionary

students={"Sumi":84, "Ram":67, "Meenu":90}
names=list(students.keys())
values=list(students.values())
plt.bar(names,values)
# Title, Label ,Grid

plt.title("BarPlot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()

```

```python
# for Horizontal Bar Plot
plt.barh(names,values,color='g')
plt.title("BarPlot")
plt.xlabel("Marks")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```
