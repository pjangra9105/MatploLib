```python
# Imports
import numpy as np
from matplotlib import pyplot as plt
```

```python
#  Defining two arrays

x=np.arange(1,11)
y=x*2

```

```python
x
```

```python
y
```

```python
# plot the x and y
plt.plot(x,y)
plt.show()
```

```python
# we can gives here Title and labels to plot
plt.title("Line_Plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

```python
# can also add the colors and linestyle and width of the Line
plt.plot(x,y,color='g', linestyle=':', linewidth=2)
plt.show()
```

```python
# we can Define two Lines with different colors and Linewidth with the Gridline
y1=2*x
y2=3*x
plt.title("Line_Plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.plot(x,y1,color='g', linestyle=':', linewidth=2)
plt.plot(x,y2,color='r', linestyle='-', linewidth=2)
plt.grid(True)
plt.show()
```

```python
# To Add Subplots
plt.subplot(1,2,1) 
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.subplot(1,2,2)
plt.plot(x,y1,color='r',linestyle='-',linewidth=3)
plt.show()
```
