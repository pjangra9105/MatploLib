```python
# Imports
from matplotlib import pyplot as plt

# creating two list

x=[22,40,20,35,67,45,80,90]
a=[1,2,3,4,5,6,7,8]

plt.scatter(x,a)
plt.show()
```

```python
# To add color,Markstyle, size in ScatterPlot
plt.scatter(x,a,c='red',marker='*',s=200)
plt.show()

```

```python
# Add TWo Marker in the Same Plot

x=[22,40,20,35,67,45,80,90]
a=[1,2,3,4,5,6,7,8]
b=[3,6,4,7,8,0,1,2]

plt.scatter(x,a,c='red',marker='*',s=200)
plt.scatter(x,b,c='green',marker='.',s=500)
plt.show()
```

```python
# To add SubPlot

plt.subplot(1,2,1)
plt.scatter(x,a,marker='*',c="g",s=300)

plt.subplot(1,2,2)
plt.scatter(x,b,marker=".",c="r",s=500)
```
